1. For each state, calculate the total number of transactions, total transaction amount, and total registered users by summing up the values from the district level data.

2. Compare the results with the corresponding values at the state level to ensure they match.

3. Display any discrepancies found between the district-level and state-level data.

```python
#3 Data quality Check
#1
district_agg = district_txn_users.groupby(
    ['State', 'Year', 'Quarter']
).agg({
    'Transactions': 'sum',
    'Amount (INR)': 'sum',
    'Registered Users': 'sum'
}).reset_index()
```

```python
state_agg = state_txn_users.groupby(
    ['State', 'Year', 'Quarter']
).agg({
    'Transactions': 'sum',
    'Amount (INR)': 'sum',
    'Registered Users': 'sum'
}).reset_index()
```

```python
comparison = pd.merge(
    state_agg,
    district_agg,
    on=['State', 'Year', 'Quarter'],
    suffixes=('_state', '_district')
)
```

```python
#2
comparison['txn_diff'] = comparison['Transactions_state'] - comparison['Transactions_district']
comparison['amount_diff'] = comparison['Amount (INR)_state'] - comparison['Amount (INR)_district']
comparison['users_diff'] = comparison['Registered Users_state'] - comparison['Registered Users_district']
```

```python
#3
discrepancies = comparison[
    (comparison['txn_diff'] != 0) |
    (comparison['amount_diff'] != 0) |
    (comparison['users_diff'] != 0)
]

print(discrepancies)
```

4.2: Correlate population density with transaction volume

1. Merge the District_Txn and Users dataset with the District Demographics dataset.

2. Calculate the correlation between population density and transaction volume.

3. Create a scatter plot to visualize the correlation between population density and transaction volume.

```python
#4.2 
print(district_demographics.columns)
```

```python
#1 
merged = pd.merge(
    district_txn_users,
    district_demographics,
    on=['State', 'District']
)
```

```python
#2
district_data = merged.groupby(
    ['State', 'District']
).agg({
    'Transactions': 'sum',
    'Density': 'first'
}).reset_index()
```

```python
correlation = district_data['Density'].corr(
    district_data['Transactions']
)

print("Correlation between Density and Transactions:", correlation)
```

```python
#3.
import matplotlib.pyplot as plt

plt.figure()

plt.figure()

# Scatter
plt.scatter(
    district_data['Density'],
    district_data['Transactions'],
    alpha=0.6
)

# 🔥 ADD TREND LINE HERE
import numpy as np

# Sort data
sorted_data = district_data.sort_values('Density')

# Fit line
z = np.polyfit(
    sorted_data['Density'],
    sorted_data['Transactions'],
    1
)
p = np.poly1d(z)

# Plot clean line
plt.plot(
    sorted_data['Density'],
    p(sorted_data['Density']),
    color='red',
    linewidth=2
)
plt.xscale('log')
plt.yscale('log')
plt.xlabel("Population Density")
plt.ylabel("Transaction Volume")
plt.title(" Population Density vs Transaction Volume")
plt.savefig("density_vs_transactions.png", dpi=300, bbox_inches='tight')
plt.tight_layout()
plt.show()
```
